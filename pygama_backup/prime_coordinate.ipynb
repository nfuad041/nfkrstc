```python
import numpy as np
from primePy import primes
from matplotlib import pyplot as plt
plt.style.use('/global/u2/f/fnafis/nfuad.mpl')
```

```python
primelist = [1] + primes.first(500-1)
print(primelist[0:10])
len(primelist)
```

```python
x = [i for i in range(10)]
print('x:',x)
y = np.polyval([2,-10],x)
print('y:',y)
plt.plot(x,y)

x_prime = [int((i/abs(i))*primelist[abs(i)]) for i in x if not i==0]
y_prime = [int((i/abs(i))*primelist[abs(i)]) for i in y if not i==0]
plt.plot(x_prime, y_prime)
print('x_prime:',x_prime)
print('y_prime:',y_prime)
```
