```python
# get file list and load energy data (numpy array)
# lh5_dir = os.path.expandvars(dg.config['lh5_dir'])
#lh5_dir = dg.lh5_dir
#dsp_list = lh5_dir + dg.fileDB['dsp_path'] + '/' + dg.fileDB['dsp_file']

#beginning_cycle = 2019

# run_0_bkg = dsp_list[run_0[0]-beginning_cycle:run_0[1]-beginning_cycle]
# #run_1_bkg = dsp_list[9:20]
# run_2_bkg = dsp_list[2028-beginning_cycle:2038-beginning_cycle]
# run_5_bkg = dsp_list[96:104]
# run_8_bkg = dsp_list[122:127]
# run_12_bkg = dsp_list[206:261]
# run_16_bkg = dsp_list[329:342]

# #runtimes are in minutes
# run_0_bkg_runtime = dg.fileDB['runtime'][0:9].sum()
# run_1_bkg_runtime = dg.fileDB['runtime'][9:20].sum()
# run_2_bkg_runtime = dg.fileDB['runtime'][2028-beginning_cycle:2038-beginning_cycle].sum()
# run_5_bkg_runtime = dg.fileDB['runtime'][96:105].sum()
# run_8_bkg_runtime = dg.fileDB['runtime'][122:128].sum()
# run_12_bkg_runtime = dg.fileDB['runtime'][206:261].sum()
# run_16_bkg_runtime = dg.fileDB['runtime'][329:342].sum()

def plot_a_run(dg, run, runtime_normalized = True, plot_known_peaks=True, save_image=False):
    beginning_cycle = 2019
    n_bins = 6000
    adc_range = (0,6000)
    fontsize = 20
    linewidth = 2
    etype = 'trapEmax'
    colors = ['b','g','r','c','y','m','k','maroon','dodgerblue','darkgoldenrod', 'gold',
             'darkslategray', 'orange']

    
    lh5_dir = dg.lh5_dir
    dsp_list = lh5_dir + dg.fileDB['dsp_path'] + '/' + dg.fileDB['dsp_file']
    files = dsp_list[run[0]-beginning_cycle:run[1]-beginning_cycle]
    edata = lh5.load_nda(files, [etype], 'ORSIS3302DecoderForEnergy/dsp')
    
    runtime = dg.fileDB['runtime'][run[0]-beginning_cycle:run[1]-beginning_cycle].sum()

    energy_uncalibrated = edata[etype] #energy_uncalibrated
    #current_pars = [-1.87482720e-07, 4.50394213e-01, -5.95154615e-01]
    current_pars = [0.431, -0.008]
    energy_calibrated = np.polyval(current_pars, energy_uncalibrated)

    hist, bins, var = pgh.get_hist(energy_calibrated, bins=n_bins, 
                                   range=(0, np.polyval(current_pars, adc_range[1])))
    if runtime_normalized == True:
        hist = hist/runtime
        var = var/runtime
    
    
    plt.figure(figsize=(24,16))
    pgh.plot_hist(hist, bins, lw=linewidth)
    if plot_known_peaks==True:
        peak_list = pd.read_csv('peak_list.txt', delimiter=' ')
        for i in range(peak_list.shape[0]):
            label = peak_list.iloc[i,0]+'='+str(peak_list.iloc[i,1])
            plt.axvline(x=peak_list.iloc[i,1], label=label, color=colors[i%12], ls='--', lw=2)
        plt.legend(bbox_to_anchor=(1, 1), loc='upper left', ncol=1, fontsize=fontsize)
    plt.yscale('log')
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    keV_per_bin = (adc_range[1]-adc_range[0])/n_bins*current_pars[0]
    plt.xlabel('trapEmax_cal (keV) (1 bin ~ '+str(keV_per_bin)+' keV)', fontsize=fontsize)
    plt.title(run[2] + ', runtime=' + str(runtime) + ' min, runtime_normalized='+
              str(runtime_normalized), fontsize = fontsize)
    if save_image: plt.savefig(run[2])
    #plt.show()
```

```python
dg = DataGroup('/global/homes/f/fnafis/krstc/krstc.json', load=True)

plot_a_run(dg, run_0, runtime_normalized=True, save_image=False)
plot_a_run(dg, run_1, runtime_normalized=True, save_image=False)
plot_a_run(dg, run_2, runtime_normalized=True, save_image=False)
```
